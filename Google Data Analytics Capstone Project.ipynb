# This python programme can automate our data cleaning task of excel csv files.

```python
import pandas as pd
from datetime import datetime

def data_cleaning(path):
    file = pd.read_csv(path, index_col=None)
    file = file.drop_duplicates()
    file = file.dropna()
    return file
```

```python
import pandas as pd
import glob

# getting excel files to be merged from the Desktop 
path = "C:\Users\Dell\Downloads\Google Data Analytics Professional Course\Course - 8 Google Data Analytics Capstone Complete a Case Study\Cyclistic Bikeshare Data"

# read all the files with extension .xlsx i.e. excel 
filenames = glob.glob(path + "\*.csv")
print('File names:', filenames)

# empty data frame for the new output excel file with the merged excel files
output_csv = pd.DataFrame()

# for loop to iterate all excel files
for file in filenames:
    # using concat for excel files
    # after reading them with read_excel()
    df = pd.concat(pd.read_excel( file, sheet_name=None), ignore_index=True, sort=False)
    # appending data of excel files
    output_csv = output_csv.append( df, ignore_index=True)

print('Final Excel sheet now generated at the same location:')
output_csv.to_csv("C:/Users/Dell/Downloads/Google Data Analytics Professional Course/Course - 8 Google Data Analytics Capstone Complete a Case Study/Cyclistic Bikeshare Data/Output.xlsx", index=False)
```
